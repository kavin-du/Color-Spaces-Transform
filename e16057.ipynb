```python
def showImages(imgs, names, width=10, height=10):
    # changing the size of canvas
    plt.rcParams['figure.figsize'] = [width, height] 

    plot = 130
    for i in range(len(imgs)):
        plot += 1
        plt.subplot(plot), plt.imshow(imgs[i]), plt.title(names[i])
    plt.tight_layout()
    plt.show()
```

```python
# custom function for converting RGB to YCrCb
def rgbToYCrCb(img):
    r,g,b = cv2.split(img)
    
    # JPEG standards for conversion
    y = np.array((0.299*r) + (0.587*g) + (0.114*b), dtype='uint8')
    cb = np.array(128 - (0.168736*r) - (0.331264*g) + (0.5*b), dtype='uint8')
    cr = np.array(128 + (0.5*r) - (0.418688*g) - (0.081312*b), dtype='uint8')

    return cv2.merge([y,cr,cb])
```

```python
    """ The YCrCb Color space"""
    
    ### Using a scale factor of 1.5 ###
    logo = cv2.cvtColor(cv2.imread('logo.png'), cv2.COLOR_BGR2RGB)
    Y,Cr,Cb = cv2.split(cv2.cvtColor(logo, cv2.COLOR_RGB2YCrCb))

    # down sampling
    factor = 1.5
    Cr_down_sample1 = cv2.resize(Cr, None, fx=1/factor, fy=1/factor)
    Cb_down_sample1 = cv2.resize(Cb, None, fx=1/factor, fy=1/factor)

    # up sampling
    Cr_up_sample1 = cv2.resize(Cr_down_sample1, (Cr.shape[1], Cr.shape[0]))
    Cb_up_sample1 = cv2.resize(Cb_down_sample1, (Cb.shape[1], Cb.shape[0]))

    temp = cv2.merge([Y, Cr_up_sample1, Cb_up_sample1])
    result1 = cv2.cvtColor(temp, cv2.COLOR_YCrCb2RGB)
```

```python
    """ Colour spaces for segmentation """
    ### segmentation in RGB space ###

    # read the image in RGB color mode
    rubic_rgb = cv2.cvtColor(cv2.imread('rubic.png'), cv2.COLOR_BGR2RGB)

    minrgb = np.array([0,118,0], dtype='uint8') # define minimum color value for green
    maxrgb = np.array([56,198,80], dtype='uint8') # define max color value for green
    mask = cv2.inRange(rubic_rgb, minrgb, maxrgb) # create the mask between color range

    # get the result
    result = cv2.bitwise_and(rubic_rgb, rubic_rgb, mask=mask)
    showImages([rubic_rgb, mask, result],['original RGB image', 'mask', 'result'])
```
