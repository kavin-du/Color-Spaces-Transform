```python
import cv2
from matplotlib import pyplot as plt
```

```python
def showImages(imgs, names, width=10, height=10):
    # changing the size of canvas
    plt.rcParams['figure.figsize'] = [width, height] 

    plot = 130
    for i in range(len(imgs)):
        plot += 1
        plt.subplot(plot), plt.imshow(imgs[i]), plt.title(names[i])
    plt.tight_layout()
    plt.show()
```

```python
""" RGB colour space """
if __name__ == '__main__':
    girl = cv2.cvtColor(cv2.imread('girl.jpg'), cv2.COLOR_BGR2RGB)
    showImages([girl], 'orginal')
```

```python
    r,g,b = cv2.split(girl)
    showImages([r,g,b], ['red', 'green', 'blue'])    
```

```python
    # merged image
    merged = cv2.merge([r,g,b])
    # merged image vs original image
    showImages([girl, merged], ['original', 'merged'])
```
