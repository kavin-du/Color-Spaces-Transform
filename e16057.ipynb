```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
def showImages(imgs, names, width=10, height=10):
    # changing the size of canvas
    plt.rcParams['figure.figsize'] = [width, height] 

    plot = 130
    for i in range(len(imgs)):
        plot += 1
        plt.subplot(plot), plt.imshow(imgs[i]), plt.title(names[i])
    plt.tight_layout()
    plt.show()
```

```python
def rgbToYCrCb(img):
    r,g,b = cv2.split(img)
    
    # JPEG standards for conversion
    y = np.array((0.299*r) + (0.587*g) + (0.114*b), dtype='uint8')
    cb = np.array(128 - (0.168736*r) - (0.331264*g) + (0.5*b), dtype='uint8')
    cr = np.array(128 + (0.5*r) - (0.418688*g) - (0.081312*b), dtype='uint8')

    return cv2.merge([y,cr,cb])
```
